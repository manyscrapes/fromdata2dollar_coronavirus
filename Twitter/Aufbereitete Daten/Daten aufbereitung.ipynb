```python
import pandas as pd
import datetime
import numpy
import string
import sys


```

```python
fin=open("tweetssommaruga.txt", "rt")
fout=open("sommaruga2.txt", "wt")

for line in fin:
    fout.write(line.replace("thisismydelimiter", ";;;"))
```

```python
fin=open("sommaruga2.txt", "rt")
fout=open("sommaruga3.txt", "wt")

for line in fin:
    fout.write(line.replace("\n", "//"))
```

```python
raw_df = pd.read_csv("sommaruga3.txt", header=None, delimiter=";;;")
raw_df
```

```python
df = raw_df.transpose()
df.head()
```

```python
clean_df = df[~df[0].str.startswith("RT", na=False)]
clean_df
```

```python
step = clean_df[0].str.split(pat = ";;", n = 1, expand = True)
step.head()
```

```python
step.dropna(inplace = True)
step.describe()
```

```python
step.columns = ["Full Text Tweet", "Datum und Uhrzeit"]
step.tail(12)
```

```python
step.describe()
```

```python
step.to_csv('sommarugaclean.csv', index=False, sep=";", encoding="utf-8-sig")
```
